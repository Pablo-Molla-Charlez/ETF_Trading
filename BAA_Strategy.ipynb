## $$ \text{STEP 3: CANARY CHANT} $$


After having understood the assets that the strategy considers, we define one of the most important functions of the strategy: $\fbox{13612W}$ Momentum Score

The strategy relies on the $\textcolor{yellow}{Canary}$ Universe to decide to which Universe we should be, either $\textcolor{green}{Defensive}$ Universe or $\textcolor{red}{Offensive}$ Universe.

The $\fbox{13612W}$ Momentum Score is defined as follows:

$$ 13612W_{ETF}(w_{1}, w_{3}, w_{6}, w_{12}) = w_{1} * (\underbrace{\frac{p_{0}-p_{1}}{p_{1}}}_{\text{1-Month Return}}) + w_{3} * (\underbrace{\frac{p_{0}-p_{3}}{p_{3}}}_{\text{3-Months Return}}) + w_{6} * (\underbrace{\frac{p_{0}-p_{6}}{p_{6}}}_{\text{6-Months Return}}) + w_{12} * (\underbrace{\frac{p_{0}-p_{12}}{p_{12}}}_{\text{12-Months Return}}) $$

where $p_{0}$ is the price of the ETF today, $p_{1}$ is the price of the ETF exactly 1-Month ago, $p_{3}$ is the price of the ETF exactly 3-Months ago and as expected $p_{12}$ is the price of the ETF exactly 12-Months ago. The weights considered are the following, $w_{1} = 12$, $w_{3} = 4$, $w_{6} = 2$ and $w_{12} = 1$.

The 13612W momentum is a more complex and less common method, specifically designed to capture absolute momentum. Absolute momentum compares the current price of an asset against its own historical prices to determine if it's generally moving up or down. It differs from relative momentum in its consideration of the asset's own performance rather than comparing it to other assets.


How the $\textcolor{yellow}{Canary}$ Universe works?

1. If [$13612W_{ETF}$] > 0 then it suggests a trend of positive performance $\longrightarrow$ we switch to the $\textcolor{red}{Offensive}$ Universe (here you choose if you want Aggressive or Balanced).

2. If [$13612W_{ETF}$] < 0 then it suggests a downtrend $\longrightarrow$ we switch to the $\textcolor{green}{Defensive}$ Universe.


If **ALL** canary assets have positive momentum, we switch to the $\textcolor{red}{Offensive}$ Universe, otherwise we remain in the $\textcolor{green}{Defensive}$ Universe until next month.

**NOTE:** Note how cautious this criterion is. If even one canary asset is exhibiting negative momentum (which has happened about 60% of the time), we shift to the $\textcolor{green}{Defensive}$ Universe.


## $$ \text{STEP 4: NEW ASSETS SELECTION} $$


We define the $\fbox{\text{Relative Momentum}}$ as follows:

$$ \text{RM}(n) = \frac{p_{0}}{average(p_{0}, ..., p_{n})} $$

where $p_{0}$ is today's price and $n$ considers the most recent $n$ month-end prices represented by $p_{i}$.

Basically, we divide todays price by the average of the $(n+1)$ last months prices. In our case, we consider $n=12$, then $13$ periods.

$\fbox{Case A}$: We switch to $\textcolor{orange}{Offensive} $ Universe

$ \fbox{\text{Case A.1:} \textcolor{red}{Aggressive} } $

We select only 1 asset out of the 4 in total, the one with the highest relative momentum.

$ \fbox{\text{Case A.2:} \textcolor{pink}{Balanced}} $

We select 6 assets out of the 12 in total, the ones with the highest relative momentum.

$\fbox{Case B}$: We stay in $\textcolor{green}{Defensive} $ Universe

We select 3 assets out of the 7 in total, the ones with the highest relative momentum. However, we consider one last case with the 3 chosen assets, $\alpha_{i} \forall i \in {1,2,3}$.

$\fbox{Case B.1}$

If $RM(\alpha_{i}) \leq RM("BIL") \longrightarrow $ The corresponding portion of the portfolio is placed in cash.

$\fbox{Case B.2}$

On the opposite, remains as the chosen asset.

$$ -------------------- \text{EXAMPLE OF HOW THE 13612 MOMENTUM WORKS} -------------------- $$



Detailed Calculation of the 13612W Momentum for the ETF "SPY"
For a specific day, let's say January 1st, 2021, you would perform the following steps:

Calculate the 1-Month Return:

Look at the closing price of SPY on January 1st, 2021 and the closing price one month earlier, for example, December 1st, 2020.
Calculate the percentage return between these two points.
Multiply this return by 12 (the weight assigned for the 1-month return).
Calculate the 3-Month Return:

Compare the closing price on January 1st, 2021 with the closing price three months earlier, say October 1st, 2020.
Calculate the percentage return between these two points.
Multiply this return by 4 (the weight for the 3-month return).
Calculate the 6-Month Return:

Similarly, calculate the percentage return between January 1st, 2021 and six months before.
Multiply this return by 2.
Calculate the 12-Month Return:

Calculate the percentage return for a full year, from January 1st, 2020 to January 1st, 2021.
Multiply this return by 1.
Add the Weighted Returns:

Now sum the four weighted returns that you calculated in the previous steps. This sum will give you the 13612W momentum for January 1st, 2021. This process is then repeated for the rest of the days in the chosen period.


$$ -------------------------- \text{END OF EXAMPLE} ------------------------- $$

## $$ \text{STEP 5: REBALANCING \& HOLDING} $$

Finally, make sure to follow the last steps:

1. Equally-weight to all assets selected at the close:

When multiple assets are selected for investment, an equal amount of capital is allocated to each asset. For example, if the strategy dictates investing in six assets, then each asset would receive one-sixth of the total capital available for investment.
"At the close" typically means that these adjustments are made at the closing price of the assets at the end of the trading day.
Rebalance the portfolio even if there isn’t a change in signal:

2. Rebalancing is the process of realigning the weightings of the assets in a portfolio. 

This ensures that the intended asset allocation is maintained over time.
This particular guideline states that the rebalancing should occur every month, even if the strategy's indicators ("signals") have not changed since the last rebalancing. This could mean buying or selling assets to maintain the equal weighting, due to changes in asset values over the month.

3. Hold all positions until the end of the following month:

Once the portfolio is set at the close of the trading day (usually on the last day of the month), all positions are held without any further buying or selling until the last day of the next month.
This rule ensures that the strategy is not reacting to market movements during the month, which could incur additional transaction costs and potentially destabilize the strategic allocation.

#### Monthly Data

```python
online_monthly = online.resample("BM").last()
online_monthly.rename(index={online_monthly.index[-1]:datetime.datetime.today()},inplace=True)
online_monthly.index = online_monthly.index.strftime('%B %d, %Y')
online_monthly
```

## Step 2: Momentum Calculation

### 13612W Momentum

```python
weights = [12, 4, 2, 1]
periods = [1,3,6,12]
# Function to compute the 13612W Momentum
def calculate_13612W_momentum(data_monthly, weights, periods):
    weighted_momentum_score = sum([(data_monthly / data_monthly.shift(month) - 1)*weight / 4 for month,weight in zip(periods, weights)]).dropna()
    return weighted_momentum_score*100


weighted_momentum_df = calculate_13612W_momentum(online_monthly, weights, periods)
```

```python
weighted_momentum_df
```

### Theory: 13612U Momentum

```python
periods = [1,3,6,12]

# Function to compute the 13612U Momentum
def calculate_13612U_momentum(data_monthly, periods):
    uniform_momentum_score = sum([(data_monthly / data_monthly.shift(month) - 1) / 4 for month in periods]).dropna()
    return uniform_momentum_score*100
        
uniform_momentum_df = calculate_13612U_momentum(online_monthly, periods)
uniform_momentum_df
```

### Theory: Relative Momentum ###

```python
# Function to compute the 13612W Momentum
def calculate_relative_momentum(data_monthly):
    # Calculate the 13 month-end prices average, including the current month (p0...p12)
    rolling_avg = data_monthly.rolling(window=13, min_periods=13).mean()

    # Calculate the relative momentum score
    relative_momentum_score = data_monthly / rolling_avg


    # Display the relative momentum score dataframe
    relative_momentum_score.dropna(inplace=True)  # Remove NaN values resulting from the shift
    
    return relative_momentum_score


relative_momentum_df = calculate_relative_momentum(online_monthly)
relative_momentum_df
```

## Step 3: Asset Selection + Canary Chant

```python
defensive_tickers = ["TIP", "DBC", "BIL", "IEF", "TLT", "LQD", "AGG"]
canary_tickers = ["SPY","EEM", "EFA", "AGG"]
offensive_aggressive_tickers = ["QQQ", "EEM", "EFA", "AGG"]
offensive_balanced_tickers = ["SPY", "QQQ", "IWM", "VGK", "EWJ", "EEM", "VNQ", "DBC", "GLD", "TLT", "HYG", "LQD"]

def canary_chant(relative_momentum):

    # Universes of assets
    offensive_aggressive_assets = relative_momentum[offensive_aggressive_tickers]
    offensive_balanced_assets = relative_momentum[offensive_balanced_tickers]
    defensive_assets = relative_momentum[defensive_tickers]
    
    # Access to the last row of Canary Assets 
    canary_assets = relative_momentum[canary_tickers].iloc[-1]
    canary_assets = pd.DataFrame(canary_assets.map(lambda x: True if x > 0 else False)).transpose()
    
    # Access to BIL Asset
    bil_asset = relative_momentum[["BIL"]].iloc[-1].iloc[0]
    
    # Access to Top Defensive Assets (Tickers and Values)
    top_defensive_assets = defensive_assets.iloc[-1].nlargest(3).index.tolist()
    top_defensive_assets_values = defensive_assets[top_defensive_assets].iloc[-1].to_list()
    
    # Assets to cash
    assets_to_cash = [(asset, asset_value < bil_asset) for asset, asset_value in zip(top_defensive_assets, top_defensive_assets_values)]

    # Cases to consider
    if all(canary_assets):
        print(f"In {canary_assets.index[0]} we switch to Offensive Universe.\n")
        print(f"    Offensive-Aggressive Asset (Top 1/4): {offensive_aggressive_assets.iloc[-1].idxmax()}")
        print(f"    Offensive-Balanced Assets (Top 6/12): {offensive_balanced_assets.iloc[-1].nlargest(6).index.tolist()}")
    else:
        print(f"In {canary_assets.index[0]} we switch to Defensive Universe.\n")
        print(f"    Defensive Assets (Top 3/7): {top_defensive_assets}")
        print(f"    Defensive Assets to cash: {assets_to_cash}")

canary_chant(relative_momentum_df)
```
